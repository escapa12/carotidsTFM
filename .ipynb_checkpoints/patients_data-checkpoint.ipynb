### Patients with Placa

```python
placa_ccd_counts=df.groupby('placas_cc_d').ones.sum()

print('Patients with ccd placa:'+str((1-placa_ccd_counts[1]/placa_ccd_counts.sum())*100)+'%')
placa_ccd_counts
```

```python
placa_cci_counts=df.groupby('placas_cc_i').ones.sum()

print('Patients with cci placa:'+str((1-placa_cci_counts[1]/placa_cci_counts.sum())*100)+'%')
placa_cci_counts
```
