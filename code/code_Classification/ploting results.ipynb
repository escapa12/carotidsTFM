```python
import numpy as np
import matplotlib.pyplot as plt

def plot_col(f,loss,val_loss,acc,val_acc):  
    axarr[f-1, 0].plot(loss, linewidth=2, label='train-loss')
    axarr[f-1,0].plot(val_loss, linewidth=2, label='val-loss')
    axarr[f-1, 0].set_title('Number of classes '+str(f*2+1)+': Loss Function')
    axarr[f-1, 0].grid()
    axarr[f-1, 0].legend()
    axarr[f-1, 0].set_xlabel('epochs')
    axarr[f-1, 0].set_yscale('log')
    
    axarr[f-1,1].plot(acc, linewidth=2, label='train-acc')
    axarr[f-1,1].plot(val_acc, linewidth=2, label='val-acc')
    axarr[f-1, 1].set_title('Number of classes '+str(f*2+1)+': Accuracy')
    axarr[f-1, 1].grid()
    axarr[f-1, 1].legend()
    axarr[f-1, 1].set_xlabel('epochs') 
```
